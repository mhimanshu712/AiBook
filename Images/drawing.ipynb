```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((512,512,3),dtype=np.int16)
```

```python
plt.imshow(img)
```

```python
cv2.rectangle(img,(300,0),(510,200),color=(0,255,0),thickness=10)
plt.imshow(img)
```

```python
# To fill thickness = -1
cv2.circle(img,(200,200),50,color=(0,0,255),thickness=10)
plt.imshow(img)
```

```python
cv2.line(img,(0,0),(100,100),(255,0,0))
plt.imshow(img)
```

```python
# Displaying text
# Remember that org = (y,x) cuz we are using origin as ref, and we are not treating it as an arr[][]

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,"Hello",(0,400),font,3,(255,255,255),10)
plt.imshow(img)
```

```python
img2 = np.zeros((512,512,3),np.int16)
# int16 did not work in vert
vert = np.array([ [50,50],[200,100],[300,500],[25,25] ],np.int32)

# copy data in all color channels
pts = vert.reshape((-1,1,2))

cv2.polylines(img2,[pts],isClosed=True,color=(23,56,100),thickness=5)
plt.imshow(img2)
```
