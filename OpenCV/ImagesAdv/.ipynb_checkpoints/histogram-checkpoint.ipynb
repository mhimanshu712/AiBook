```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

dark_horse = cv2.imread('../DATA/horse.jpg')
show_horse = cv2.cvtColor(dark_horse,cv2.COLOR_BGR2RGB)

rainbow = cv2.imread('../DATA/rainbow.jpg')
show_rainbow = cv2.cvtColor(rainbow,cv2.COLOR_BGR2RGB)

blue_bricks = cv2.imread('../DATA/bricks.jpg')
show_bricks = cv2.cvtColor(blue_bricks,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(show_horse)
```

```python
plt.imshow(show_rainbow)
```

```python
plt.imshow(show_bricks)
```

```python
# Doing hist for blue channel,mask is if you wnat to hist for a part of img
# 
hist_values = cv2.calcHist([blue_bricks],channels=[0],mask=None,histSize=[256],ranges=[0,256]) 

plt.plot(hist_values)
```

```python
# Doing hist for blue channel,mask is if you wnat to hist for a part of img
# 
hist_values = cv2.calcHist([dark_horse],channels=[0],mask=None,histSize=[256],ranges=[0,256]) 

plt.plot(hist_values)
```

```python
img = blue_bricks
color =('b','g','r')

for i,col in enumerate(color):
    histr=cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])
    
plt.title('HISTROGAM FOr BRICK')
```

```python
# That horse image is very large(and dark ovio), and so shows mostly dark in hist

img = dark_horse
color =('b','g','r')

for i,col in enumerate(color):
    histr=cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,50])
    plt.ylim([0,500000])
    
plt.title('HISTROGAM FOr HORSE')
```
